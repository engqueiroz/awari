# Gera√ß√£o de Hip√≥teses

## Hip√≥tese 1

As caracteristicas das especies s√£o bons discriminantes ( Preditores) da categoria da flor?

### Anova em todas as caracteristicas

```python
for caracteristica in df_iris.columns[:-1]:
    
    anova, pvalor = stats.f_oneway(df_iris.loc[(df_iris['especie']=='setosa'),caracteristica],
                                   df_iris.loc[(df_iris['especie']=='versicolor'),caracteristica],
                                   df_iris.loc[(df_iris['especie']=='virginica'),caracteristica])
    
    print(f"Para a caracteristica {caracteristica}: Anova: {anova}, p-value: {pvalor}")
```

### Testando todos os cenarios

```python
from itertools import permutations, combinations
```

```python
[x for x in combinations(['setosa','virginica','versicolor'],2)]
```

```python
for caracteristica in df_iris.columns[:-1]:
    
    resultado_teste = []
    
    for esp1, esp2 in combinations(['setosa','virginica','versicolor'],2):
        _, pvalue = stats.ttest_ind(df_iris.loc[df_iris['especie']==esp1, caracteristica],
                                    df_iris.loc[df_iris['especie']==esp2, caracteristica])
        
        if pvalue < 0.05:
            resultado_teste.append(True)
        else:
            resultado_teste.append(False)
    
    if sum(resultado_teste) == 3:
        print(f'A caracteristica {caracteristica} √© um bom preditor')
    else:
        print(f'A caracteristica {caracteristica} n√£o √© um bom preditor')
```

## Testes de m√©dias n√£o-param√©trico

```python
#Kolmogorov-smirnov
```

```python
for caracteristica in df_iris.columns[:-1]:
    
    resultado_teste = []
    
    for esp1, esp2 in combinations(['setosa','virginica','versicolor'],2):
        _, pvalue = stats.ks_2samp(df_iris.loc[df_iris['especie']==esp1, caracteristica],
                                    df_iris.loc[df_iris['especie']==esp2, caracteristica])
        
        if pvalue < 0.05:
            resultado_teste.append(True)
        else:
            resultado_teste.append(False)
    
    if sum(resultado_teste) == 3:
        print(f'A caracteristica {caracteristica} √© um bom preditor')
    else:
        print(f'A caracteristica {caracteristica} n√£o √© um bom preditor, {resultado_teste}')
```
