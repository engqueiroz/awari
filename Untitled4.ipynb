```python
#Crie um programa que use a função input() ( pesquise como ela funciona ) para receber dois numeros inteiros.
#armazenar seus valores em variáveis após, calcule qual é o resto da divisão entre eles.
#Após encontrar o resto da divisão, valide se esse resto é par ou impar.
#Realize uma validação para que os inputs sejam sempre numeros para que seja possivel realizar o cálculo esperado
#Para todos os calculos acima, printe as etapas e valores para cada um deles.

VAR = int(input('insert an int number '))
LET = int(input('insert another int number '))
r = VAR/LET
print('o resto é:' ,r)
if r %2 ==0: print('it is PAR')
else: print('it is IMPAR')
```

```python
#Utilizando os operadores de atribuição, crie um programa que receba um valor numérico e calcule:
#Um quinto do valor ( ⅕ )
#Valor elevado a 2 ( x² ) 
#Retorna “Par” se o número for par e “Impar” se o numero for impar

ex3 = float(input('insert a number: '))
print('the number is: ' +str(ex3))
ex3q = ex3
ex3q /=5
ex3e = ex3
ex3e **= 2
print('um quinto do numero é: '+ str(ex3q))
print('a potencia de 2 do numero é: ' +str(ex3e))
if ex3 %2 == 0: print(str(ex3) + ' é numero par')
else: print(str(ex3) + ' é numero impar')
```
