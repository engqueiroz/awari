```python
"""while cont < contagem: """

def exclusao (contagem, dataframe):
    cont = 0
    for i in range(contagem):
        dataframe.drop(cont, axis = 0, inplace=True)
        cont += 1
        
    return dataframe
```

```python
df_teste_delete = df.copy()
df_delete = exclusao(10, df_teste_delete)
df_delete.head(20)
```
